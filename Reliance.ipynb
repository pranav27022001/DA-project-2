```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as exp
import plotly.graph_objects as go
```

```python
df = pd.read_csv('reliance_data.csv')
```

```python
df.shape
```

```python
df['Last'].fillna(df['Last'].mean(), inplace=True)
df['Trades'].fillna(df['Trades'].mean(), inplace=True)
df['Deliverable Volume'].fillna(0, inplace=True)
df['%Deliverble'].fillna(0, inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["Date"]=pd.to_datetime(df["Date"], format="%d-%m-%Y")
df["Date"]
df.set_index('Date', inplace=True)
df.info()
```

```python
df.head()
```

```python
plt.figure(figsize=(16,8))
plt.title('Reliance Industries Close Price History')
plt.plot(df['Close'])
plt.xlabel('Date',fontsize=18)
plt.ylabel('Close Price INR',fontsize=18)
plt.show()
```

Above is the analysis done for close price of reliance stock from year 1996-2020.

```python
plt.figure(figsize=(16,8))
plt.title('Reliance Industries Turn Over from 1996 to 2020')
plt.plot(df['Turnover'])
plt.xlabel('Date',fontsize=18)
plt.ylabel('TurnOver of Reliance INR',fontsize=18)
plt.show()
```

From above analysis we figured out the turnover of Reliance from 1996-2020. Highest turnover for reliance was in year 2020 which was nearly 14 billion.

```python
plt.figure(figsize=(16,8))
plt.title('Reliance all time High from 1996 to 2022')
plt.plot(df['High'])
plt.xlabel('Date',fontsize=18)
plt.ylabel('High of Reliance INR',fontsize=18)
plt.show()
```

From above analysis we have figured out the all time high price of Reliance between year 1996-2020. The all time high of reliance was near to 3500.

```python
plt.figure(figsize=(16,8))
plt.title('Reliance Trades from 1996 to 2020')
plt.plot(df['Trades'])
plt.xlabel('Date',fontsize=18)
plt.ylabel('High of Reliance INR',fontsize=18)
plt.show()
```

Above are the highest trades of Reliance from year 1996 to 2022. The maximum trade count is in year 2020

```python
candlestick = go.Candlestick(
                            x=data_2020.index,
                            open=data_2020['Open'],
                            high=data_2020['High'],
                            low=data_2020['Low'],
                            close=data_2020['Close']
                            )
fig = go.Figure(data=[candlestick])

fig.update_layout(
    width=1200, height=600,
    title="Reliance, 2020",
    yaxis_title='Price'
)

fig.show()
```
